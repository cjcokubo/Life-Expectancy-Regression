```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/Users/xtinejoy/Downloads/Regression Dataset.csv")
df.head()
```

```python
df_2015 = df[df['Year'] == 2015.0]
df_2015.head()
```

```python
df_2015.columns.values
```

```python
regression_data = df_2015[['Schooling', 'Life expectancy']]
regression_data = regression_data.dropna()
```

```python
regression_data.plot(kind='scatter', x='Schooling', y='Life expectancy')
plt.show()
```

```python
def best_fit(X, Y):
    xbar = sum(X)/len(X)
    ybar = sum(Y)/len(Y)
    n = len(X) # or len(Y)

    numer = sum([xi*yi for xi,yi in zip(X, Y)]) - n * xbar * ybar
    denum = sum([xi**2 for xi in X]) - n * xbar**2

    b = numer / denum
    a = ybar - b * xbar

    print('best fit line:\ny = {:.5f} + {:.5f}x'.format(a, b))

    return a, b

a, b = best_fit(regression_data['Schooling'], regression_data['Life expectancy'])
```

```python
X = regression_data['Schooling']
Y = regression_data['Life expectancy']

plt.scatter(X, Y)
yfit = [a + b * xi for xi in X]
plt.plot(X, yfit)
plt.xlabel('Schooling')
plt.ylabel('Life expectancy')
plt.show()
```
